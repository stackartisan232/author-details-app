```python
 !pip install requests beautifulsoup4

import requests
from bs4 import BeautifulSoup
import csv
```

```python
quotes = []
page = 1

while True:
  url = f"https://quotes.toscrape.com/page/{page}/"
  response = requests.get(url)

  soup = BeautifulSoup(response.text,"html.parser")
  quote_element = soup.find_all("div", class_="quote")

  if not quote_element:
    print("No more quotes. Stopping....")
    break

  for quote in quote_element:
    text = quote.find("span", class_="text").get_text()
    author = quote.find("small", class_="author").get_text()
    quotes.append([text, author])

  print("Scraped Page:", page)
  page += 1

print("Total quotes collected:",
      len(quotes))

```

```python
import csv

with open("quotes.csv", "w", newline="", encoding="utf-8") as file:
  writer = csv.writer(file)
  writer.writerow(["Quote", "Author"])

  writer.writerows(quotes)

print("Saved to quotes.csv")
```
