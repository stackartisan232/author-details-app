```python
import pandas as pd
import matplotlib.pyplot as plt

quotes = [
   "The world as we have created it as a process of our thinking.",
    "It is our choices ,Harry, that shows what we truly are, far more than our abilities.",
   "There is no greater agony than bearing an untold story inside you.",
   "Be the change that we wish to see in the world.",
   "Imperfection is a beauty, madness is genius.",
   "Try not to become a man of success, but rather than try to be a man of value.",
   "It's better to be hated for what you are than to be loved for what you are not.",
   "A person who never made mistake never learned new."
   ]

authors = [
       "Albert Einstein",
       "J.K Rowling",
       "Maya Angelou",
       "Mahatma Gandhi",
       "Marilyn Monroe",
       "Albert Einstein",
       "Andre gide",
       "Thomas A. Edison"
       ]

categories = [
      "thinking",
      "choices",
      "pain",
      "Change",
      "beauty",
      "success",
      "truth",
      "mistakes",
  ]

df = pd.DataFrame({
       "Quote": quotes,
       "Author": authors,
       "Category": categories
   })

print("Dataset preview")
print(df)
print("\ncolumns:",  df.columns.tolist())

author_counts = df["Author"].value_counts()
print("\nNumber of quotes per author:")
print(author_counts)


plt.figure(figsize=(8, 5))
author_counts.plot(kind="bar")
plt.title("Number of Quotes per Author")
plt.xlabel("Author")
plt.ylabel("count")
plt.tight_layout()
plt.show()
```
