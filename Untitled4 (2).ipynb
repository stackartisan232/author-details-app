```python
import random
import time

def slow_print(text, delay=0.05):
    for char in text:
        print(char, end='', flush=True)
        time.sleep(delay)
    print()

slow_print("Welcome to the adventure game.")
player_name = input("Enter you name: ")
slow_print("Hello {player_name}, your journey begins now...\n")

player_health = 100
inventory = []

slow_print("you find yourself at a crossroad. where do you want to go?")
slow_print("1.Enter the dark forest.")
slow_print("2. walk towards the village.")
choice1 = input("Enter 1 or 2: ")

if choice1 == "1":
  slow_print("\nYou enter the dark forst. It's spooky! ")
  encounter = random.choice(["monster", "treasure", "nothing"])

  if encounter == "monster":
    slow_print("a monster appears and attacks you!")
    fight_choice = input("Do you want to fight or run?(fight/run): ")
    if fight_choice.lower() == "fight":
      damage = random.randit(10, 30)
      player_health -= damage
      slow_print(f"you fought bravely but lost{damage} HP. your health is now {player_health}.")
    else:
      slow_print("you run away safely.\n")

  elif encounter == "treasure":
   slow_print("you found a hidden treasure chest!")
   treasure = random.choice(["Magic sword", "Healing potion",])
   inventory.append(treasure)
   slow_print(f"you got a {treasure}!\n")

  else:
    slow_print("Nothing happens. You continue walking...\n")

else:
  slow_print("\nyou walk towards the village and meet friendly villagers.")
  slow_print("They give you some food and a map!")
  inventory.append("Map")
  slow_print(f"Inventory: {inventory}\n")

slow_print("After a while, you reach a raging river. You need to cross it.")
slow_print("options:")
slow_print("1. swim across")
slow_print("2. look for a bridge")
choice2 = input("Enter 1 or 2: ")

if choice2 == "1":
  slow_print("\nyou decide to swim across the river....")
  if "Magic sword" in inventory:
    slow_print("Your magic sword glows and protects you! you cross safely.\n")
  else:
    damage = random.randit(10, 25)
    player_health -= damage
    slow_print(f"The current is strong! You lost {damage} HP. Your health is now {player_health}.")
else:
  slow_print("You search for a bridge and find one downstream. You cross safely.\n")

slow_print("You arrive at a cave rumored to have a dragon guarding treasure.")
slow_print("Do you want to enter? (yes/no)")
choice3 = input().lower()

if choice3 == "yes":
  slow_print("You enter the cave and encounter a dragon!")
  if "Magic sword" in inventory:
    slow_print("You use the magic sword and defeat the dragon!")
    inventory.append("Dragon Treasure")
  else:
    damage = random.randint(20, 40)
    player_health -=damage
    slow_print(f"you are injured by the dragon! Lost {damage} HP. Health: {player_health}")
    slow_print("you escape the cave safely but empty handed.\n")
else:
  slow_print("You decide it's too dangerous and walk safely.\n")

slow_print(f"\n Adventure completed, {player_name}! ")
slow_print(f"Health: {player_health}")
slow_print(f"Inventory: {inventory}")

if "Dragon Treasure" in inventory:
  slow_print("congratulations! you are a legendary hero and found the dragon's treasure!")
elif player_health > 50:
  slow_print("you survived the adventure with your health intact. Well done!")
else:
  slow_print("you survived, but it was a tough journey. Better luck next time!")



```
